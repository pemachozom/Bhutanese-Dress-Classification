```python
from sklearn.model_selection import train_test_split
import os

# Define the paths to your data
data_dir = "/content/drive/MyDrive/new/train"
inappropriate_dir = os.path.join(data_dir, "inappropriate")
appropriate_dir = os.path.join(data_dir, "appropriate")

# List the image files in each category
inappropriate_images = [os.path.join(inappropriate_dir, file) for file in os.listdir(inappropriate_dir)]
appropriate_images = [os.path.join(appropriate_dir, file) for file in os.listdir(appropriate_dir)]

# Define labels for your data (0 for inappropriate, 1 for appropriate)
inappropriate_labels = [0] * len(inappropriate_images)
appropriate_labels = [1] * len(appropriate_images)

# Combine the data and labels
all_images = inappropriate_images + appropriate_images
all_labels = inappropriate_labels + appropriate_labels

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(all_images, all_labels, test_size=0.2, random_state=42)

# Now you have X_train, X_test, y_train, and y_test that you can use for training and testing your model.
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# Create an ImageDataGenerator with desired augmentations
datagen = ImageDataGenerator(
    rotation_range=40,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest'
)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense

model = Sequential()
model.add(Conv2D(32, (3, 3), activation='relu', input_shape=(150, 150, 3)))
model.add(MaxPooling2D((2, 2)))
model.add(Conv2D(64, (3, 3), activation='relu'))
model.add(MaxPooling2D((2, 2)))
model.add(Conv2D(128, (3, 3), activation='relu'))
model.add(MaxPooling2D((2, 2)))
model.add(Flatten())
model.add(Dense(128, activation='relu'))
model.add(Dense(1, activation='sigmoid'))

model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```
